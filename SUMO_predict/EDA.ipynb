```python
# 重複データの削除
df_1_1 = df.drop_duplicates()
df_1_1.reset_index(drop=True, inplace=True)
df_1_1.duplicated().sum()
```

```python
df_1_2 = df_1_1
```

### 1.2 データ型変換

- 'access' は複数の駅からの距離平均を表す変数を作る．int型．
- 'age' は築年数を抽出．int型．
- 'floor' はint型に変換
- 'fee' はint型に変換
- 'management_fee' はint型に変換
- 'deposit' はint型に変換
- 'gratuity' はint型に変換．
- 'madori' は部屋数をint型に変換．LDKの区分をカテゴリカル変数にする．
- 'mennseki' はfloat型に変換． 

####  'access'の処理

```python
import re
```

```python
access_mean = []
for i in range(0, len(df_1_2)):
    s = df_1_2['access'].iloc[i]
    result = re.findall(r"\d+", s)
    result = list(map(int, result))
    result_mean = np.mean(result)
    access_mean.append(result_mean)
```

```python
df_1_2['access_mean'] = access_mean
```

#### 'age'の処理

```python
age = []
for i in range(0, len(df_1_2)):
    s = df_1_2['age'].iloc[i]
    if re.search('新築', s):
        age.append('0')
    else:
        result = re.findall(r"\d+", s)
        age.append(result[0])

age = list(map(int, age))
```

```python
df_1_2['age'] = age
df_1_2.head()
```

#### 'floor'の処理

```python
floor = []
for i in range(0, len(df_1_2)):
    s = df_1_2['floor'].iloc[i]
    result = re.findall(r"\d+", s)
    result = list(map(int, result))
    floor.append(result)
```

```python
df_1_2['floor'] = floor
df_1_2.head()
```

#### 'fee'の処理

```python
fee = []
for i in range(0, len(df_1_2)):
    s = df_1_2['fee'].iloc[i]
    result = re.findall(r"[0-9]+\.?[0-9]", s)
    result = list(map(float, result))
    result = result*10000
    print(result)
    fee.append(result)
```
