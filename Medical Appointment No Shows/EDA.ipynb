#### 欠損値の確認

```python
def calc_missing_rate(df, v):
    total=df[v].isnull().sum()
    percent = round(total/len(df[v])*100,2)
    
    return pd.DataFrame([[total, percent]], columns=['total','percent'], index=[v])
```

```python
for i in df.columns:
    display(calc_missing_rate(df, i))
```

#### カテゴリー変数の変換

```python
# one-hot encoding
df_droped = pd.get_dummies(df_droped, columns=['Gender','No-show'])
```

```python
# label encoding

label = df_droped['Neighbourhood'].unique()

le = LabelEncoder()
le.fit(label)

df_droped = le.transform(df_droped)
```
